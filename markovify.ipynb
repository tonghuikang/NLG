```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import markovify # Markov Chain Generator
from IPython.display import display, HTML # to display-print dataframe inline
```

```python
import glob
files = []
for file in glob.glob("./*.csv"):
    print(file)
    files.append(file)
```

```python
df_list = []

for file in files:
    df = pd.read_csv(file)
    display(df.sample(n=5))
    df_list.append(df)
```

# show result trained on one corpus

```python
# selecting the first corpus
df = df_list[0]
```

```python
comments = [str(i) for i in df.Comment]
# apparently "nan" in list([...,"nan",...]) screws things up

text_model = markovify.NewlineText(comments, state_size=2)
# state size could not go above 3 :/

# Print ten randomly-generated sentences using the built model
for i in range(10):
    print(text_model.make_sentence())
```

# show result trained on respective corpus

```python
for file,df in zip(files,df_list):
    
    comments = [str(i) for i in df.Comment]
    # apparently "nan" in list([...,"nan",...]) screws things up

    text_model = markovify.NewlineText(comments, state_size=2)
    # state size could not go above 3 :/
    
    print("\n\n", file, "\n")

    # Print ten randomly-generated sentences using the built model
    for i in range(8):
        print(text_model.make_sentence())
```
