```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import markovify # Markov Chain Generator
from IPython.display import display, HTML # to display-print dataframe inline
import time
```

```python
import glob
files = []
for file in glob.glob("./*.csv"):
    print(file)
    files.append(file)
```

```python
df_list = []

for file in files:
    df = pd.read_csv(file)
    display(df.sample(n=5))
    df_list.append(df)
```

# show result trained on one corpus

```python
# selecting the first corpus
df = df_list[0]
```

```python
comments = [str(i) for i in df.Comment]
# apparently "nan" in list([...,"nan",...]) screws things up

start_time = time.time()
text_model = markovify.NewlineText(comments, state_size=2)
print("Time taken : ", round(time.time() - start_time,2), "s")
# state size could not go above 3 :/

# Print ten randomly-generated sentences using the built model
for i in range(10):
    print(text_model.make_sentence())
```

# show result trained on respective corpus

```python
for file,df in zip(files,df_list):
    
    comments = [str(i) for i in df.Comment]
    # apparently "nan" in list([...,"nan",...]) screws things up

    start_time = time.time()
    text_model = markovify.NewlineText(comments, state_size=2)
    print("\n\n", file, "\n", "Time taken : ", round(time.time() - start_time,2), "s")
    # state size could not go above 3 :/
    
    # Print ten randomly-generated sentences using the built model
    for i in range(8):
        print(text_model.make_sentence())
```

# Trying out AashitaK's POSifiedText
It is still not a comment because it does not have contextual references :(

```python
import spacy
nlp = spacy.load("en_core_web_lg")
```

```python
class POSifiedText(markovify.Text):
    def word_split(self, sentence):
        return ["::".join((word.orth_, word.pos_)) for word in nlp(sentence)]

    def word_join(self, words):
        sentence = " ".join(word.split("::")[0] for word in words)
        return sentence
```

```python
comments = [str(i) for i in df.Comment][:1000]
```

```python
start_time = time.time()
next_model_POSified = POSifiedText(comments, state_size = 2)
print("Time taken : ", round(time.time() - start_time,2), "s")
```

```python
for i in range(10):
    print(next_model_POSified.make_sentence())
```
